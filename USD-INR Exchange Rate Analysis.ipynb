```python
query1 = """
Select  date_format(str_to_date(Date, '%%m-%%d-%%Y'), '%%M %%d,%%Y') as formated_date, value
FROM crude_oil_price
"""
df = pd.read_sql(query1, engine)
df
```

```python
df.to_csv(r"C:\Users\chari\Downloads/cleaned_crude_oil_price.csv", index= False)
```

```python
query2 = """
Select  date_format(str_to_date(Date, '%%d-%%m-%%Y'), '%%M %%d,%%Y') as formated_date, open, high, low, close
FROM usdinr
"""
df1 = pd.read_sql(query2, engine)
df1
```
